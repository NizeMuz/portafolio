```python
import mysql.connector
#propiedad ---- NizeMuz

conexion = mysql.connector.connect(user="redsocial_admin", password="admin_12",
                                   host="localhost",
                                   database="redsocial")




def menu():
    print("MENU--------------------------")
    print("1-. Agregar usuario")
    print("2.- Modificar la contrase単a")
    print("3.- Probar inicio de usuario")
    print("4.- Salir")
    opcion = int(input("Ingrese una opcion"))
    return opcion

opcion = menu()

def agregar():
    cursor = conexion.cursor()
    agregar_usuario = ("INSERT INTO usuarios "
                    "(nombre, email, contrasena) "
                    "VALUES (%s, %s, %s)")

    nombre = input("Ingrese el nomnbre del usuario")
    email = input("Ingresar el email del usuario")
    contrasena = input("Ingrese la contrase単a del usuario")

    datos_usuario = (nombre, email, contrasena)

    cursor.execute(agregar_usuario, datos_usuario)

    conexion.commit()
    cursor.close()
    print("Datos guardados")

def modificar():
    cursor = conexion.cursor()

    cmd_consulta = ("SELECT idUsuario, nombre, email, contrasena "
                "FROM usuarios")

    cursor.execute(cmd_consulta)

    for (idUsuario, nombre, email, contrasena) in cursor:
        print(idUsuario, "-", nombre, "-", email, "-", contrasena)

    emailUsuario = input("ingrese el email del usuario que desea modificar")

    cmd_modificar = ("UPDATE usuarios "
                    "SET contrasena = %s"
                    "WHERE email = %s")

    contrasena = input("Ingrese la nueva contrasena para el usuario seleccionado")
    
    valores = (contrasena, emailUsuario)

    cursor.execute(cmd_modificar, valores)
    conexion.commit()
    cursor.close()
    print("Datos guardados")

def validacion_datos():
    print("Ingrese su Mail y contrase単a")
    usuario = input("Ingrese su Mail")
    pasword = input("Ingrese su contrase単a")

    cursor = conexion.cursor()
    consulta =("SELECT email,contrasena FROM usuarios WHERE email=%s AND contrasena=%s")


    cursor.execute(consulta,(usuario,pasword))
    resultado = cursor.fetchall()

    if resultado:
        for i in resultado:
            print("Bienvenido "+i[0])

    else:
        print("El usuario o la contrasena son incorrectas")
        intentarotravez = input("Quiere intentarlo otra vez? s/n:")
        if intentarotravez.lower() == "n":
            print("Adios")
            
    conexion.commit()
    cursor.close()
    print("Datos verificados")    



while opcion != 4:
    if opcion == 1:
        agregar()

    if opcion == 2:
        modificar()
    
    if opcion == 3:
        validacion_datos()

    opcion = menu()

print("Ha salido")

conexion.close()
```
